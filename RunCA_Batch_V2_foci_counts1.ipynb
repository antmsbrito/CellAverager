```python
def distance(p1,p2):
    # l2 norm
    return np.sqrt((p1[0]-p2[0])**2+(p1[1]-p2[1])**2)

def minpointdistance(points_1,points_2):
    
    mind = 1e12
    for i in points_1:
        for j in points_2:
            if i!=j:
                d = distance(i,j)
                if d<mind:
                    mind=d
    if mind!=1e12:
        return mind*0.08
    else:
        return None
```
